#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
titanic.Age.hist()
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
fig,ax = plt.subplots(1,3)
bins_number = [10,20,50]
for i in range(3):
    plt.sca(ax[i])
    titanic.Age.hist(bins=bins_number[i])
```

#### Whether the bin size affects your plot? Why?

```python
# Yes, because the more bins are selected the more categories there are for a given data point to fall into, so it gives more or less information on the plot
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. Choose the appropriate plot to show below.

```python
plt.boxplot(titanic.Age)
```

#### What does the previous plot tell you about the feature `Age`? Comment.

```python
# It tells us the median, as well as Q1 and Q3 values. It also calculates a range using the IQR and shows us the outliers
```

#### Now in addition to the summary statistics, we also want to visualize the distribution of `Age`. Choose the appropriate plot to sow below.

```python
plt.violinplot(titanic.Age, showmedians=True, quantiles=[0.25,.5,.75])
```

```python
sns.violinplot(x="Age", data=titanic)
```
